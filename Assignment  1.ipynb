```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("used_cars_data.csv")

plt.hist(df["Price"].dropna(), bins=20, color="skyblue", edgecolor="black")

plt.xlabel("Price (Lakhs)")
plt.ylabel("Number of Cars")
plt.title("Used Car Price Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

brands = ['Lamborghini', 'Bentley', 'Porsche', 'Jaguar', 'Land', 
          'Mini', 'Audi', 'Mercedes-Benz', 'BMW', 'Isuzu']
avg_prices = [4.8, 4.1, 3.7, 3.5, 3.3, 3.2, 3.1, 3.0, 3.0, 2.9] 
plt.bar(brands, avg_prices, color='pink', edgecolor='black')
plt.title("Brand Vs Price")
plt.xlabel("Brand")
plt.ylabel("Average Price (Lakhs)")
plt.xticks(rotation=45)

plt.show()
```
